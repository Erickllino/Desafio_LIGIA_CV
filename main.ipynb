```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

## 1. Análise dos dados

```python
df = pd.read_csv('data/training.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
df.describe()
```

Então temos 30 keypoints e 1 uma imagem cada linha

## 2. Pre-Processamento dos dados

```python
# Como as imagens estão em formato de string, precisamos convertê-las para um array numpy
# Na descrição do dataset, é dito que as imagens são 96x96 pixels, então vamos usar isso para redimensionar as imagens

df['Image'] = df['Image'].apply(lambda x: np.fromstring(x, sep=' ').reshape(96, 96))
```

```python
df['Image'].head()
```

```python
plt.figure(figsize=(10, 6))
# mostrar a imagem do dataframe vendo os valores como uma imagem
print(df['Image'][2].shape)
plt.imshow(df['Image'][2], cmap='grey')
plt.axis('off')
plt.show()
```

## 3. Separando os dados

```python
X_train = np.array(df['Image'].tolist())
y_train = df.drop(columns=['Image']).values
```

## 4. Modelo

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
```

```python
model = models.Sequential()
model.add(layers.Conv2D(3, (3, 3), activation='relu', input_shape=(96, 96, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(6, (3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(30, activation='sigmoid'))

model.summary()
```

```python
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['RootMeanSquaredError', 'accuracy'])

history = model.fit(X_train, y_train, epochs=50, batch_size=128, validation_split=0.2)
```

```python
model.predict(X_train[0:1].reshape(1, 96, 96, 1))
# plotar a imagem original e os keypoints previstos
plt.figure(figsize=(10, 6))
plt.imshow(X_train[0], cmap='gray')
plt.scatter(model.predict(X_train[0:1].reshape(1, 96, 96, 1))[0][0:30:2], model.predict(X_train[0:1].reshape(1, 96, 96, 1))[0][1:30:2], c='red')
plt.scatter(y_train[0][0:30:2], y_train[0][1:30:2], c='blue')
plt.axis('off')
plt.show()
```
